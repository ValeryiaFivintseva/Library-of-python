Задание 1

```python
import pandas as pd
import numpy as np
authors = pd.DataFrame({'author_id': ['1', '2', '3'], 'author_name': ['Тургенев', 'Чехов', 'Островский']},
                       columns = ['author_id', 'author_name'])
authors
```

```python
book = pd.DataFrame({'author_id': ['1', '1', '1', '2', '2', '3', '3'],
                     'book_title': ['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза',
                                    'Таланты и поклонники'], 'price': [450, 300, 350, 500, 450, 370, 290]},
                   columns = ['author_id', 'book_title', 'price'])
book
```

Задание 2

```python
authors_price = authors.merge(book, on='author_id')
authors_price.set_index('author_id', inplace = True)
authors_price
```

Задание 3

```python
top5 = authors_price.nlargest(5, 'price')
top5
```

Задание 4

```python
price = authors_price.groupby('author_name').agg({'price': 'min'})
price = price.rename(columns = {'price': 'min_price'})
price2 = authors_price.groupby('author_name').agg({'price': 'max'})
price2 = price2.rename(columns = {'price': 'max_price'})
mn_price = authors_price.groupby('author_name').agg({'price': 'mean'})
mn_price = mn_price.rename(columns = {'price': 'mean_price'})
authors_stat = price.merge(price2, on = 'author_name') \
.merge(mn_price, on = 'author_name')
authors_stat
```

Задание 5**

```python
authors_price['cover'] = ['твердая', 'мягкая', 'мягкая', 'твердая', 'твердая', 'мягкая', 'мягкая']
book_info = pd.pivot_table(authors_price, index = ['author_name'], values = ['price'], columns = ['cover'],
                           aggfunc = [np.sum], fill_value = 0)
book_info
```
